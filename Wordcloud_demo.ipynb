```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Create a list of word
text = ("Python Python Python Matplotlib Matplotlib Seaborn DataScience BusinessAnalyst DataAnalyst Plot Violin Numpy Pandas SPYDER JUPYTER NLP NLTK NLTK NLTK NLP NOTEBOOK PYCHARM VSCODE ABC ")
```

```python
text
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=700, height=400, margin=2, background_color='black', colormap='Accent', mode='RGBA').generate(text)
```

```python
# Display the generated image:
plt.imshow(wordcloud, interpolation='quadric')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
# Display the generated image:
plt.imshow(wordcloud, interpolation='spline36')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
